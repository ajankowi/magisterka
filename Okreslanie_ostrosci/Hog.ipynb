```python
import dlib
from imageio import imread
import matplotlib.pyplot as plt
import cv2
import os
import dlib
```

```python
directory = r'/home/adam_j4/magisterka/Okreslanie_ostrosci/Ostre'  

image_files = [file for file in os.listdir(directory) if file.endswith('.jpg') or file.endswith('.png')]

ostre = []
for file in image_files:
    image_path = os.path.join(directory, file)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    ostre.append(image)
```

```python
# Create a HOG face detector using the built-in dlib class
face_detector = dlib.get_frontal_face_detector()
```

```python
detected_faces = []

# Run the HOG face detector on the image data
for i in range(len(ostre)):
    # Create a HOG face detector using the built-in dlib class
    face_detector = dlib.get_frontal_face_detector()

    # Run the HOG face detector on the image data
    detected_faces.append(face_detector(ostre[i], 1))

# Loop through each list of rectangles in the detected_faces list
for i, face_rects in enumerate(detected_faces):
    # Loop through each rectangle in the list
    
    for face_rect in face_rects:

        # Draw a box around each face
        plt.figure(figsize=(4, 4))
        plt.imshow(ostre[i][face_rect.top():face_rect.bottom(), face_rect.left():face_rect.right()], cmap='gray')
        plt.show()
```
