```python
import dlib
from imageio import imread
import matplotlib.pyplot as plt
import cv2
```

```python
path = '/home/adam_j4/magisterka/Okreslanie_ostrosci/Rozmyte/5.png'

image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
```

```python
# Create a HOG face detector using the built-in dlib class
face_detector = dlib.get_frontal_face_detector()
```

```python
# Run the HOG face detector on the image data
detected_faces = face_detector(image, 1)
```

```python
print("Found {} faces in the image file".format(len(detected_faces)))

# Loop through each face we found in the image
for i, face_rect in enumerate(detected_faces):

    # Draw a box around each face we found
    plt.figure(figsize=(4,4))
    plt.imshow(image[face_rect.top():face_rect.bottom(), face_rect.left():face_rect.right()], cmap='gray')
    plt.show()
```

```python
import dlib
from imageio import imread
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/home/adam_j4/magisterka/Okreslanie_ostrosci/Rozmyte/1.png', cv2.IMREAD_GRAYSCALE)

# Load the model
model_path = "/home/adam_j4/magisterka/Okreslanie_ostrosci/mmod_human_face_detector.dat"  # replace with the path to your model
cnn_face_detector = dlib.cnn_face_detection_model_v1(model_path)

# Run the face detector on the image data
detected_faces = cnn_face_detector(image, 1)

print("Found {} faces in the image file".format(len(detected_faces)))

# Loop through each face we found in the image
for i, face_rect in enumerate(detected_faces):

    # Draw a box around each face we found
    plt.figure(figsize=(4,4))
    plt.imshow(image[face_rect.rect.top():face_rect.rect.bottom(), face_rect.rect.left():face_rect.rect.right()], cmap='gray')
    plt.show()
```
