```python
import glob
import imageio
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
from tensorflow.keras import layers
import time
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

from IPython import display
```

```python
#Wczytanie zdjęć treningowych (
images_train = []
labels_train = []
size = 70

for i in range(1,size+1):
    im = Image.open('../../Inz/mnist/Praca_inz/Zdjecia/Baza_zdjec_konwersja_64x64/s10/'+str(i)+'.png')
    arr_im = np.array(im)
    arr_im = (arr_im - 127.5) / 127.5  # Normalize the images to [-1, 1]
    images_train.append(arr_im)
    labels_train.append(0) 

    
for j in range(1,size+1):
    im = Image.open('../../Inz/mnist/Praca_inz/Zdjecia/Baza_zdjec_konwersja_64x64/s2/'+str(i)+'.png')
    arr_im = np.array(im)
    arr_im = (arr_im - 127.5) / 127.5  # Normalize the images to [-1, 1]
    images_train.append(arr_im)
    labels_train.append(1) 

print(len(images_train))
```

```python
a = 60
plt.imshow(images_train[a],cmap = 'gray')
print(labels_train[a])
```

```python
# Zakładając, że masz już przygotowane dane obrazów i etykiet
BUFFER_SIZE = 140
BATCH_SIZE = 32

# Tworzenie zestawu danych z obrazów i etykiet
train_dataset_with_labels = tf.data.Dataset.from_tensor_slices((images_train, labels_train))

# Tasowanie i dzielenie danych na partie
train_dataset = train_dataset_with_labels.shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
```

```python
def make_generator_model():
    input_latent = tf.keras.Input(shape=(100,))
    input_label = tf.keras.Input(shape=(1,))

    # Embedding etykiety i jej przekształcenie do wektora
    label_embedding = layers.Embedding(10, 100)(input_label)
    label_embedding = layers.Flatten()(label_embedding)

    # Łączenie wektora latentnego i etykiety
    merged_input = layers.Concatenate()([input_latent, label_embedding])

    x = layers.Dense(16*16*256, use_bias=False)(merged_input)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Reshape((16, 16, 256))(x)

    x = layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    x = layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    x = layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)

    x = layers.Conv2DTranspose(1, (5, 5), strides=(1, 1), padding='same', use_bias=False, activation='tanh')(x)

    model = tf.keras.Model(inputs=[input_latent, input_label], outputs=x)
    return model
```

```python
def make_discriminator_model():
    input_image = tf.keras.Input(shape=(64, 64, 1))
    input_label = tf.keras.Input(shape=(1,))

    # Embedding etykiety i jej przekształcenie do odpowiedniego rozmiaru
    label_embedding = layers.Embedding(10, 64 * 64)(input_label)
    label_embedding = layers.Reshape((64, 64, 1))(label_embedding)

    # Łączenie obrazu z przekształconą etykietą
    concatenated_inputs = layers.Concatenate(axis=-1)([input_image, label_embedding])

    x = layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same')(concatenated_inputs)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(256, (5, 5), strides=(2, 2), padding='same')(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(1)(x)

    model = tf.keras.Model(inputs=[input_image, input_label], outputs=x)
    return model
```

```python
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator = make_generator_model()
discriminator = make_discriminator_model()

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)

@tf.function
def train_step(images, labels):
    noise = tf.random.normal([tf.shape(labels)[0], 100])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator([noise, labels], training=True)

        real_output = discriminator([images, labels], training=True)
        fake_output = discriminator([generated_images, labels], training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
```

```python
EPOCHS = 20


for epoch in range(EPOCHS):
    start_time = time.time()
    for image_batch, label_batch in train_dataset:
        train_step(image_batch, label_batch)
    end_time = time.time()
    epoch_duration = end_time - start_time
    print(f"Czas trwania epoki {epoch + 1}: {epoch_duration:.2f} sekund")
    
    
    # Zapisanie modelu co 10 epok
    if (epoch + 1) % 10 == 0:
        generator.save(f'./Generator/generator_model_epoch_{epoch + 1}.h5')
        discriminator.save(f'./Discriminator/discriminator_model_epoch_{epoch + 1}.h5')
        print(f'Modele zapisane po {epoch + 1} epokach.')

# Zapisanie modelu po zakończeniu treningu
generator.save('./Generator/generator_model_final.h5')
discriminator.save('./Discriminator/discriminator_model_final.h5')
```

```python
def generate_and_save_images(model_path, label, num_images=10, save_path=None):
    
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Wczytywanie zapisanego modelu
    model = tf.keras.models.load_model(model_path)
    
    # Generowanie losowych wektorów latentnych
    noise = tf.random.normal([num_images, 100])
    
    # Tworzenie etykiet
    labels = np.array([label] * num_images).reshape(-1, 1)
    
    print(labels)
    
    # Generowanie obrazów
    generated_images = model([noise, labels], training=False)
    
    # Normalizacja obrazów z [-1, 1] do [0, 1]
    generated_images = (generated_images + 1) / 2.0
    
    # Tworzenie wykresu dla wygenerowanych obrazów
    fig, axs = plt.subplots(1, num_images, figsize=(num_images * 2, 2))
    for i in range(num_images):
        axs[i].imshow(generated_images[i, :, :, 0], cmap='gray')
        axs[i].axis('off')
        
        if save_path:
            img = Image.fromarray((generated_images[i, :, :, 0] * 255).numpy().astype(np.uint8))
            img.save(f"{save_path}/generated_image_{i}.png")
    
    plt.show()
```

```python

# Przykład użycia funkcji
generate_and_save_images('./Generator/generator_model_final.h5', label=0, num_images=10, save_path="./images")
```
